V Percentile:

To calculate percentile, you take the time-series and arrange it in order from highest to lowest, then you see what number of days the implied volatility was lower that it is today, and divide it by the total number of days. If 80% of the days the NIFTY had an implied volatility below 7.7%, then the percentile would be 80%.

#### IVpercentile = NumDaysBelowIVtoday/TotalNumDays

```python
def final_dataframe(a):
    chain = options.get_options_chain(a)

    call = chain['calls']
    df1 = pd.DataFrame.from_dict(call)
    put = chain['puts']
    df2 = pd.DataFrame.from_dict(put)

    df1['Implied Volatility']
        
    pd.set_option('display.max_rows',181)
    df = pd.concat([df1, df2],ignore_index=True)
    df = pd.concat([df1, df2],ignore_index=True)
    list = []
    for i in df['Implied Volatility']:
        j = i.split('%')
        list.append(j[0])
        df['IV'] = pd.DataFrame(list)
    df.to_csv(r'C:\Users\YASHA\Desktop\files\file.csv')
    df = pd.read_csv(r'C:\Users\YASHA\Desktop\files\file.csv')
    IV_mean = (df['IV'].sum()/len(df['IV']))
    list2 = []
    for i in range(len(df['IV'])):
        list2.append(IV_mean)
        df['IV_mean'] = pd.DataFrame(list2)
    counts = df['IV'].value_counts()
    plt.figure(figsize=(20,10))
    df['IV'].plot(linestyle='-', markevery=IV_mean, marker='o', markerfacecolor='black')
    df['IV_mean'].plot(linestyle = '-')
    print('IV_mean : %f' %(IV_mean))
    print('IV_counts: ', counts)
    fig = plt.figure(figsize=(10,21))
    ax2= sns.countplot(y="IV",data=df)
    
    return 
```

### 1) Yellow line indicates the IV mean and Blue line provides the Implied Volatility in first plot
### 2) Countplot which can be used for IV percentile
